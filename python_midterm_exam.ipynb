## Question 1: String cleanup from a trade headline (10 points)

You are given a short headline-like sentence.

1. Convert the text to lowercase.
2. Split it into words.
3. Remove commas and periods from the ends of words.
4. Count how many times the word `"tariffs"` appears.
5. Join the cleaned words back into one string separated by a single space.

```python
headline = "Trump weighs tariffs on China, while tariffs on steel and batteries remain under debate."

# Your answer here

headline = headline.lower()
print(headline)

words = headline.split()
print(words)

cleaned_words = []
for word in words:
    cleaned_word = word.strip(",")
    cleaned_word = cleaned_word.strip(".")
    cleaned_words.append(cleaned_word)
print(cleaned_words)

counter = 0
for words in cleaned_words:
    if words == "tariffs":
        counter += 1
print(f'The word "tariffs" appears {counter} times in the headline.')

joined_headline = " ".join(cleaned_words)
print(joined_headline)
```

```python
tariff_data = {
    "solar": {"rates": [25, 30, 35], "strategic": True},
    "steel": {"rates": [10, 15, 20], "strategic": True},
    "apparel": {"rates": [5, 8, 6], "strategic": False},
    "semiconductors": {"rates": [18, 22, 25], "strategic": True}
}

# Your answer here
tariff_data_averaged = {}
for product, details in tariff_data.items():
    rates = round(sum(details["rates"]) / len(details["rates"]))
    strategic = details["strategic"]
    tariff_data_averaged[product] = {"average_rate": rates, "strategic": strategic}
    print(f"{product.capitalize()} tariffs: Rates - {rates}, Strategic - {strategic}")

for product, details in tariff_data_averaged.items():
    max_rate = max(tariff_data[product]["rates"])
print(f'the highest average tariff rate is {max_rate} for {product}.')

for product, details in tariff_data_averaged.items():
    if details["average_rate"] > 20:
        tariff_data[product]["high_tariff"] = True
    else:
        tariff_data[product]["high_tariff"] = False
print(tariff_data)
```

## Question 4: Sets and overlaps in sanction lists (8 points)

Two policy teams each built a list of countries appearing in separate sanctions memos.

Tasks:
1. Convert both lists into sets.
2. Find the countries that appear in both lists.
3. Find the countries unique to Team A.
4. Print `"Team B covered more unique countries"` if Team B has more unique entries than Team A. Otherwise print `"Team A covered at least as many unique countries"`.

```python
team_a = ["Russia", "Iran", "China", "Russia", "Belarus", "Iran"]
team_b = ["China", "North Korea", "Iran", "Syria", "China", "Russia", "Venezuela"]

# Your answer here

team_a_set = set(team_a)
team_b_set = set(team_b)

common_countries = team_a_set.intersection(team_b_set)
print(f"Countries in both teams: {common_countries}")

team_b_set = team_b_set - common_countries
team_a_set = team_a_set - common_countries
if len(team_b_set) > len(team_a_set):
    print("Team B has more unique countries.")
elif len(team_a_set) > len(team_b_set):
    print("Team A has more unique countries.")
else:
    print("Both teams have the same number of unique countries.")

#went down a rabbit hole of len(list) vs list.len and object oriented coding in chatgpt. Apparently you can do __len__() as well.
team_a_set.__len__()
print(f"Team A has {len(team_a_set)} unique countries.")
```

## Question 6: Recursion with nested trade flows (12 points)

Write a recursive function called `sum_trade_flows(data)` that takes a nested list of integers and returns the sum of all integers inside it.

Example:
- `sum_trade_flows([100, [50, 25], [10, [5, 5]], 20])` should return `215`

Assume the list may contain integers or other lists.

After writing the function, test it on:
`flows = [100, [50, 25], [10, [5, 5]], 20]`

```python
flows = [100, [50, 25], [10, [5, 5]], 20]

# Your answer here

def sum_trade_flows(data):
    total = 0
    for item in data:
        if isinstance(item, list):
            total += sum_trade_flows(item)
        else:
            total += item
    return total

sum_of_flows = sum_trade_flows(flows)
print(f"The total sum of trade flows is: {sum_of_flows}")
```

## Question 7: Try / except with messy commodity-price data (12 points)

You are given a list of values that are supposed to represent oil prices, but some entries are invalid.

Tasks:
1. Loop through the list and try to convert each value to a float.
2. Use `try` / `except` to skip invalid entries.
3. Store valid numeric values in a new list.
4. Compute the average of the valid values.
5. Print how many invalid entries were skipped.

```python
oil_prices = ["82.4", "79.1", "N/A", "", "95.0", "spike", None, "101.3", "88.2"]

# Your answer here

def clean_oil_prices(prices):
    cleaned_prices = []
    for price in prices:
        try:
            cleaned_price = float(price)
            cleaned_prices.append(cleaned_price)
        except (ValueError, TypeError):
            continue
    return cleaned_prices

cleaned_prices = clean_oil_prices(oil_prices)
print(f"Cleaned oil prices: {cleaned_prices}")
```
